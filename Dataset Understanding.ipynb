```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import streamlit as st
```

```python
#Load in and print the dataset as well as check for null values and data types.
```

```python
dataset = pd.read_csv('World-Stock-Prices-Dataset.csv')
```

```python
print(dataset.info())
dataset.isnull().sum()
```

```python
#Start the Cleaning of the data 
```

```python
dataset["Date"] = pd.to_datetime(
dataset["Date"],
utc = True,)
```

```python
dataset = dataset.drop_duplicates()
```

```python
dataset.to_csv("cleaned_stock_data.csv", index=False)
```

```python
clean_data = pd.read_csv("cleaned_stock_data.csv")
print(clean_data.info())
```

```python
#Data has been cleaned and organized now we can move on to the next step of the project which is data visualization.
```

```python
#Data Visualizations
```

```python
apple = clean_data[clean_data["Ticker"] == "AAPL"]

fig = px.line(
    apple,
    x="Date",
    y="Close",
    title="Apple Stock Price"
)

fig.show()
```

```python
fig = px.bar(
    apple,
    x="Date",
    y="Volume"
)

fig.show()
```
